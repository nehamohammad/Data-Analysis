show the score comparison of 2 students in 5 different subjects,subjects as x-axis,marks as y-axis

```python

#plot the scores of 2 students in 5 different subjects, subjects as x-axis and marks as y axis
stud1=[10,20,24,35,50]
stud2=[12,34,25,44,48]
sub=["telugu","hindi","english","maths","science"]
plt.plot(sub,stud1,label="student1",color="blue")
plt.plot(sub,stud2,label="student2",color="black")
plt.title("marks vs subjects")
plt.legend()
```

```python
stud1=[10,20,24,35,50]
stud2=[12,34,25,44,48]
sub=["telugu","hindi","english","maths","science"]
plt.subplot(2,1,1)#subplot(rows,columns,positions)
plt.plot(sub,stud1,label="student1",color="blue")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,stud2,label="student2",color="black")
plt.title("marks vs subjects")
plt.legend()
```

```python
revA=np.array([230,560,780,127,128])
revB=np.array([200,160,270,127,400])
profitC=np.diff(revA)
print(profitC)
profitD=np.diff(revB)
print(profitD)
year=["2019-2020","2020-2021","2021-2022","2022-2023"]
plt.subplot(1,2,1)
plt.bar(year,profitC,label="company 1",color="black")
plt.legend()
plt.subplot(1,2,2)
plt.bar(year,profitD,label="company 2",color="green")
plt.legend()
```

Pie chart


```python
#syntax:pie(values,labels=label_name)
#n(values)=n(labels)
A=np.array([25,40,35,10])
labe=["python","AIML","Cloud","numpy"]
plt.pie(A,labels=labe)
```

```python
A=np.array([25,40,35,10])
labe=["python","AIML","Cloud","numpy"]
explo=[0.2,0,0,0]
plt.pie(A,labels=labe,explode=explo,startangle=180)
plt.show()
```

```python
A=np.array([25,40,35,10])
labe=["python","AIML","Cloud","numpy"]
explo=[0.2,0,0,0]
plt.pie(A,labels=labe,explode=explo,startangle=180,shadow=True)
plt.legend()
plt.show()
```
